### Exercise 1: Line Plot Customization

```python
# TODO: Create a line plot with x values ranging from 0 to 10 and y values as the square of x.
# Customize the plot by adding a title, labels for both axes, and a grid.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate x values ranging from 0 to 10
x = np.linspace(0, 10, 100)

# Calculate y values as the square of x
y = x ** 2

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='y = x^2', color='blue')

# Add title and axis labels
plt.title('Line Plot of y = x^2', fontsize=16)
plt.xlabel('x', fontsize=14)
plt.ylabel('y', fontsize=14)

# Add a grid
plt.grid(True, linestyle='--', alpha=0.7)

# Add a legend
plt.legend(fontsize=12)

# Show the plot
plt.show()
```
