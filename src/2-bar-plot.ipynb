### Exercise 2: Bar Plot with Colors

```python
# TODO: Create a bar plot with the following data: categories = ['A', 'B', 'C', 'D'] and values = [5, 7, 3, 9].
# Use different colors for each bar and add a title to the plot.
```

```python
import matplotlib.pyplot as plt

# Create a bar plot with the specified data
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
colors = ['red', 'green', 'blue', 'orange']

plt.figure(figsize=(8, 6))
plt.bar(categories, values, color=colors)

# Add title to the bar plot
plt.title('Bar Plot of Categories and Values', fontsize=16)
plt.xlabel('Categories', fontsize=14)
plt.ylabel('Values', fontsize=14)

# Show the bar plot
plt.show()
```
